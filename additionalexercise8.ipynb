```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_borrowed = False

    def borrow(self):
        if self.is_borrowed:
            return f"'{self.title}' is already borrowed."
        self.is_borrowed = True
        return f"You borrowed '{self.title}'."

    def return_book(self):
        if not self.is_borrowed:
            return f"'{self.title}' was not borrowed."
        self.is_borrowed = False
        return f"You returned '{self.title}'."


class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author):
        self.books.append(Book(title, author))

    def find_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                return book
        return None

    def show_books(self):
        for book in self.books:
            status = "Borrowed" if book.is_borrowed else "Available"
            print(f"{book.title} by {book.author} - {status}")


# Test Run
library = Library()
n = 2

library.add_book("Python 101", "Guido")
library.add_book("OOP Basics", "John Doe")

print(library.find_book("Python 101").borrow())
print(library.find_book("Python 101").return_book())
library.show_books()
```

```python
class Employee:
    def __init__(self, name, rate, hours):
        self.name = name
        self.rate = rate
        self.hours = hours

    def calculate_salary(self):
        if self.hours <= 160:
            return self.rate * self.hours, 0

        overtime_hours = self.hours - 160
        base_salary = self.rate * 160
        overtime_pay = overtime_hours * self.rate * 1.5
        return base_salary, overtime_pay


# Test Run
employees = []

employees.append(Employee("Alice", 100, 170))
employees.append(Employee("Bob", 120, 150))

for emp in employees:
    base, overtime = emp.calculate_salary()
    print(f"{emp.name}")
    print(f"Hours Worked: {emp.hours}")
    print(f"Base Salary: {base}")
    print(f"Overtime Pay: {overtime}")
    print("---")
```

```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock


class Cart:
    def __init__(self):
        self.items = []

    def add_item(self, product):
        if product.stock <= 0:
            return f"{product.name} is out of stock."

        self.items.append(product)
        product.stock -= 1
        return f"{product.name} added to cart."

    def remove_item(self, product_name):
        for product in self.items:
            if product.name == product_name:
                self.items.remove(product)
                product.stock += 1
                return f"{product_name} removed from cart."
        return "Item not found in cart."

    def total_price(self):
        return sum(item.price for item in self.items)

    def show_cart(self):
        for item in self.items:
            print(f"{item.name} - {item.price}")
        print("Total:", self.total_price())


# Test Run
p1 = Product("Laptop", 1000, 5)
p2 = Product("Mouse", 50, 10)

cart = Cart()

print(cart.add_item(p1))
print(cart.add_item(p2))
print(cart.remove_item("Mouse"))
cart.show_cart()
```
