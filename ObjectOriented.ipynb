```python
# Step 1: Define class with encapsulation
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.__balance = balance  # Private attribute

    def deposit(self, amount):
        self.__balance += amount
        return f"New balance: ${self.__balance}"

    def get_balance(self):
        return f"Balance: ${self.__balance}"

# Step 2: Create object
account = BankAccount("Alice", 5000)

# Step 3: Test methods
print(account.deposit(1500))
print(account.get_balance())
```

```python
# Step 1: Base class
class Person:
    def __init__(self, name):
        self.name = name

    def introduce(self):
        return f"Hi, I'm {self.name}."

# Step 2: Subclass (Inheritance)
class Student(Person):
    def introduce(self):
        return f"Hi, I'm {self.name}, and I'm a student."

# Step 3: Create objects
person = Person("Allen")
student = Student("Mark")

# Step 4: Call methods
print(person.introduce())
print(student.introduce())
```

```python
# Step 1: Base class
class Shape:
    def area(self):
        return "Calculating area..."

# Step 2: Subclass Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

# Step 3: Subclass Rectangle
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Step 4: Create objects
circle = Circle(5)
rectangle = Rectangle(4, 6)

# Step 5: Print results
print(f"Circle area: {circle.area()}")
print(f"Rectangle area: {rectangle.area()}")
```

```python
# Step 1: Define class with static method
class MathOperations:
    @staticmethod
    def square(num):
        return num * num

# Step 2: Get user input
user_input = int(input("Enter a number to square: "))

# Step 3: Call static method
result = MathOperations.square(user_input)

# Step 4: Print result
print("Square of", user_input, ":", result)
```

```python
# Step 1: Define class with operator overloading
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def __add__(self, other):
        return BankAccount(self.balance + other.balance)

# Step 2: Create objects
account1 = BankAccount(5000)
account2 = BankAccount(3000)

# Step 3: Add accounts using +
total_balance = account1 + account2

# Step 4: Print result
print(f"Total Balance: P{total_balance.balance}")
```

```python
from abc import ABC, abstractmethod

# Step 1: Abstract class
class Appliance(ABC):
    @abstractmethod
    def turn_on(self):
        pass

# Step 2: Subclass implementing abstract method
class WashingMachine(Appliance):
    def turn_on(self):
        return "Washing machine is now running."

# Step 3: Create object
wm = WashingMachine()

# Step 4: Call method
print(wm.turn_on())
```

```python
# Step 1: Define classes
class A:
    def show(self):
        return "Class A"

class B(A):
    def show(self):
        return "Class B"

class C(A):
    def show(self):
        return "Class C"

class D(B, C):
    pass  # MRO decides which show() is used

# Step 2: Create object
obj = D()

# Step 3: Call method
print(obj.show())
```

```python
class Vehicle:        # 👴 Grandparent class
    def move(self):
        return "Vehicle is moving"

class Car(Vehicle):   # 👨 Parent class
    def fuel_type(self):
        return "Uses gasoline"

class ElectricCar(Car):  # 👶 Child class (3rd level)
    def fuel_type(self):
        return "Uses electricity"

# Create object
tesla = ElectricCar()

# Call methods
print(tesla.move())
print(tesla.fuel_type())
```

```python
class ATM:
    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):
        if amount > self.balance:
            return "Insufficient balance"

        self.balance -= amount
        return f"Withdrawn {amount}, Remaining balance: {self.balance}"

# Create object
atm = ATM(1000)

# Test withdrawal
print(atm.withdraw(200))
```

```python
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title):
        self.books.append(title)
        return f"Book '{title}' added to the library"

    def show_books(self):
        return f"Available Books: {', '.join(self.books)}"

# Create object
library = Library()

# Add book
print(library.add_book("Python Programming"))

# Show books
print(library.show_books())
```
